```python
from sentence_transformers import SentenceTransformer
sentence_model =SentenceTransformer("all-MiniLM-L6-v2")
```

```python
from umap import UMAP
umap_model = UMAP(n_neighbors=50, n_components=10, min_dist=0.1, metric='cosine', random_state=101)


```

```python
from hdbscan import HDBSCAN
hdbscan_model = HDBSCAN(min_cluster_size=75, metric='euclidean', cluster_selection_method='eom', prediction_data=True)

```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer_model = CountVectorizer(min_df =2, max_df = 10 ,stop_words="english")



from bertopic.vectorizers import ClassTfidfTransformer
ctfidf_model = ClassTfidfTransformer()

```

```python
topic_model = BERTopic(
    embedding_model=sentence_model,
    umap_model=umap_model,
    hdbscan_model=hdbscan_model,
    vectorizer_model=vectorizer_model,
    ctfidf_model=ctfidf_model,
    language="english",
    calculate_probabilities=True,
    verbose=True,
    nr_topics = 300
)


topics, probs = topic_model.fit_transform(docs)
```

```python
import re

def preprocess_corpus(corpus):

    preprocessed_corpus = []
    for doc in corpus:
        # Tokenize the document into individual words
        words = doc.split()

        # Remove 'tump' and 'said'
        words = [word for word in words if word.lower() not in ['tump', 'said']]

        # Rejoin the words back into a document
        preprocessed_doc = ' '.join(words)
        preprocessed_corpus.append(preprocessed_doc)

    return preprocessed_corpus
```

```python
docs_no_trump = preprocess_corpus(docs)
```

```python
hdbscan_model = HDBSCAN(min_cluster_size=35, metric='euclidean', cluster_selection_method='eom', prediction_data=True)
```

```python
topic_model2 = BERTopic(
    embedding_model=sentence_model,
    umap_model=umap_model,
    hdbscan_model=hdbscan_model,
    vectorizer_model=vectorizer_model,
    ctfidf_model=ctfidf_model,
    language="english",
    calculate_probabilities=True,
    verbose=True,
    nr_topics = 20
)

topics, probs = topic_model2.fit_transform(docs_no_trump)
```
