```python
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
import nltk

def tokenize_and_stem(df):
    stemmer = PorterStemmer()

    processed_df = df.copy()
    def process_text(text):
        # Tokenize
        tokens = word_tokenize(str(text).lower())

        # Stem
        stemmed_tokens = [stemmer.stem(token) for token in tokens]

        return {
            'tokens': tokens,
            'stemmed_tokens': stemmed_tokens,
            'processed_text': ' '.join(stemmed_tokens)
        }

    processed_data = [process_text(text) for text in processed_df['content']]

    # Add new columns to DataFrame
    processed_df['tokens'] = [data['tokens'] for data in processed_data]
    processed_df['stemmed_tokens'] = [data['stemmed_tokens'] for data in processed_data]
    processed_df['processed_text'] = [data['processed_text'] for data in processed_data]

    return processed_df
```

```python
processed_df = tokenize_and_stem(df)
```
